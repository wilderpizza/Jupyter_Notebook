### The code writed below allows you to calculate your BMI (Body Mass Index)

```python
def bodyMassIndex(w, h): 
   
    return w/h**h

w = float(input("Introduce your weight in KG: "))
h = float(input("Introduce your height in meters: "))

bmi = bodyMassIndex(w, h) 
print("Your BMI is {}".format(bmi))

#it works in python 3
```
